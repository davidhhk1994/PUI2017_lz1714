```python
# fetch and open the data from MTA using MTA_KEY
apiKey = os.getenv("mtaKEY")
url = "http://bustime.mta.info/api/siri/vehicle-monitoring.json?key=" + apiKey + "&VehicleMonitoringDetailLevel=calls&LineRef=B63"
```

```python
response = urllib.urlopen(url)
data = response.read()
data = json.loads(data)
```

```python
content_siri_serv_vehi = data['Siri']['ServiceDelivery']['VehicleMonitoringDelivery']
```
